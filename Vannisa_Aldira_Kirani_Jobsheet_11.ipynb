Praktikum 01 - Sequential Search pada Data Angka

```python
def sequential_search(data, target):
  for i in range(len(data)):
    if data[i] == target:
      return f"Data {target} ditemukan di indeks {i}"
  return f"Data {target} tidak ditemukan"

angka = [10, 20, 30, 40, 50]
print("Data:", angka)
print(sequential_search(angka, 30))
print(sequential_search(angka, 100))
```

Praktikum 02 - Binary Search pada Data Angka

```python
def binary_search(data, target):
  left, right = 0, len(data)-1
  while left <= right:
      mid = (left + right) // 2
      if data[mid] == target:
           return f"Data {target} ditemukan di indeks {mid}"
      elif data[mid] < target:
        left = mid + 1
      else:
        right = mid - 1
  return f"Data {target} tidak ditemukan"

angka = [10, 20, 30, 40, 50]
print("Data:", angka)
print(binary_search(angka, 40))
print(binary_search(angka, 100))
```

Praktikum 03 - Sequential Search untuk Data Nama

```python
def sequential_search_nama(data, target):
  for i in range(len(data)):
    if data[i].lower() == target.lower():
      return f"Nama {target} ditemukan di indeks {i}"
  return f"Nama {target} tidak ditemukan"

nama = ["Andi", "Budi", "Cici", "Doni"]
print("Data:", nama)
print(sequential_search_nama(nama, "Cici"))
print(sequential_search_nama(nama, "Eka"))
```

Praktikum 05 - Perbandingan Waktu Sequential vs Binary Search

```python
import time

def sequential_search(data, target):
  for i in range(len(data)):
    if data[i] == target:
      return i
  return -1

def binary_search(data, target):
  left, right = 0, len(data)-1
  while left <= right:
    mid = (left + right) // 2
    if data[mid] == target:
      return mid
    elif data[mid] < target:
      left = mid + 1
    else:
      right = mid - 1
  return -1

data = list(range(1, 100001))
target = 99999

start = time.time()
sequential_search(data, target)
end = time.time()
print("Sequential Search Waktu:", end - start, "detik")

start = time.time()
binary_search(data, target)
end = time.time()
print("Binary Search Waktu:", end - start, "detik")
```

Praktikum 06 - Kombinasi Sorting dan Searching

```python
import random

def insertion_sort(data):
  for i in range(1, len(data)):
    key = data[i]
    j = i - 1
    while j >= 0 and data [j] > key:
      data[j + 1] = data[j]
      j -= 1
    data[j + 1] = key
  return data

def binary_search(data, target):
  left, right = 0, len(data)-1
  while left <= right:
    mid = (left + right) // 2
    if data[mid] == target:
      return f"{target} ditemukan di indeks {mid}"
    elif data[mid] < target:
      left = mid + 1
    else:
      right = mid - 1
  return f"{target} tidak ditemukan"

data = [random.randint(1, 100) for _ in range(15)]
print("Data acak sebelum diurutkan:", data)
urut = insertion_sort(data)
print("Data setelah diurutkan:", urut)
target = urut[7]
print(binary_search(urut, target))
```

Tugas Variasi Sequential Search

```python
def sequential_search(data, target):
  count = 0
  for i in range(len(data)):
    count += 1
    if data[i] == target:
      return f"Data {target} ditemukan di indeks {i} setelah {count} perbandingan"
  return f"Data {target} tidak ditemukan setelah {count} perbandingan"

angka = [10, 20, 30, 40, 50]
print("Data:", angka)
print(sequential_search(angka, 30))
print(sequential_search(angka, 100))
```

Tugas Variasi Binary Search

```python
def binary_search(data, target):
  left, right = 0, len(data)-1
  langkah = 0
  while left <= right:
      langkah += 1
      mid = (left + right) // 2
      if data[mid] == target:
          return f"Data {target} ditemukan di indeks {mid} setelah {langkah} langkah"
      elif data[mid] < target:
          left = mid + 1
      else:
          right = mid - 1
  return f"Data {target} tidak ditemukan setelah {langkah} langkah"

angka = [10, 20, 30, 40, 50]
print("Data:", angka)
print(binary_search(angka, 40))
print(binary_search(angka, 100))
```

Tugas Studi Kasus Barang Toko

```python
def sort_barang_by_harga(data):
  n = len(data)
  for i in range(n):
      for j in range(0, n - i - 1):
          if data[j][1] > data[j + 1][1]:
                data[j], data[j + 1] = data[j + 1], data[j]
  return data

def binary_search_harga(data, target):
  left, right = 0, len(data) - 1
  while left <= right:
      mid = (left + right) // 2
      harga_mid = data[mid][1]
      if harga_mid == target:
          return f"Barang dengan harga {target}: {data[mid][0]}"
      elif harga_mid < target:
          left = mid + 1
      else:
          right = mid - 1
  return f"Tidak ada barang dengan harga {target}"

barang = [
    ("Buku", 5000),
    ("Pensil", 3000),
    ("Pulpen", 3500),
]

print("Daftar barang sebelum diurutkan:")
print(barang)

barang_urut = sort_barang_by_harga(barang.copy())

print("\nDaftar barang setelah diurutkan berdasarkan harga (asc):")
print(barang_urut)

print("\nHasil pencarian (berdasarkan harga):")
print(binary_search_harga(barang_urut, 5000))
print(binary_search_harga(barang_urut, 1000))
```

Tugas Analisis Kompleksitas

```python
import random
import time

def sequential_search(data, target):
  for i in range(len(data)):
    if data[i] == target:
      return i
  return -1

def binary_search(data, target):
  left, right = 0, len(data)-1
  while left <= right:
    mid = (left + right) // 2
    if data[mid] == target:
      return mid
    elif data[mid] < target:
      left = mid + 1
    else:
      right = mid - 1
  return -1

ukuran = [100, 1000, 10000]

print("Analisis Perbandingan Waktu Sequential dan Binary Search")

for n in ukuran:
  data = [random.randint(1, 100000) for _ in range(n)]

  target = data[-1]

  start = time.time()
  sequential_search(data, target)
  waktu_seq = time.time() - start

  data_sorted = sorted(data)
  start = time.time()
  binary_search(data_sorted, target)
  waktu_bin = time.time() - start

  print(f"\nJumlah data : {n}")
  print(f"Waktu Sequential Search : {waktu_seq:.6f} detik")
  print(f"Waktu Binary Search     : {waktu_bin:.6f} detik")
```

```python
def sort_mahasiswa_by_nilai(data):
  n = len(data)
  for i in range(n):
    for j in range(0, n - i - 1):
      if data[j][1] < data[j + 1][1]:
        data[j], data[j + 1] = data[j + 1], data[j]
  return data

def binary_search_nilai(data, target):
  left, right = 0, len(data) - 1
  while left <= right:
    mid = (left + right) // 2
    nilai_mid = data[mid][1]
    if nilai_mid == target:
      return f"Nilai {target}: {data[mid][0]}"
    elif nilai_mid > target:
      left = mid + 1
    else:
      right = mid - 1
  return f"Tidak ada mahasiswa dengan nilai {target}"

mahasiswa = []
jumlah = int(input("Berapa banyak mahasiswa? "))

for i in range(jumlah):
  nama = input(f"Nama mahasiswa ke-{i+1}: ")
  nilai = int(input("Nilai: "))
  mahasiswa.append((nama, nilai))

print("\nDaftar mahasiswa:")
print(mahasiswa)

nilai_urut = sort_mahasiswa_by_nilai(mahasiswa.copy())

print("\nDaftar mahasiswa setelah diurutkan berdasarkan nilai (desc):")
print(nilai_urut)

print("\nHasil pencarian (berdasarkan nilai):")
print(binary_search_nilai(nilai_urut, 88))
print(binary_search_nilai(nilai_urut, 70))
```
