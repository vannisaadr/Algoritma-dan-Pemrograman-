Tugas Modul Butan Sendiri

```python
# kode untuk modul buatan sendiri
```

```python
%%writefile konversi.py
def celcius_farenheit(c):
    print("Hasil konversi Celcius ke Fahrenheit =", (9/5) * c + 32)

def fahrenheit_celcius(f):
    print("Hasil konversi Fahrenheit ke Celcius =", 5/9 * (f - 32))
```

```python
import konversi

konversi.celcius_farenheit(20)
konversi.fahrenheit_celcius(80)
```

Tugas Studi Kasus Mini

```python
# kode untuk tugas studi kasus mini
```

```python
%%writefile laporan.py
def ringkasan(benar, total):
    print(f"\nSkor akhir: {benar}/{total}")
```

```python
import random
import laporan

total = 5
benar = 0

for _ in range(total):
    a = random.randint(1,10)
    b = random.randint(1,10)
    jawaban = int(input(f"{a} + {b} = "))

    if jawaban == a + b:
        benar += 1

laporan.ringkasan(total, benar)
```
