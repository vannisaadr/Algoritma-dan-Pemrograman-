Praktikum 04 - Fungsi dengan *args dan **kwargs

```python
#kode untuk praktikum 04
```

```python
def jumlahkan(*args):
  return sum(args)

def biodata(**kwargs):
  for key, value in kwargs.items():
    print(f"{key}: {value}")

print("Total:", jumlahkan(1, 2, 3, 4, 5))
biodata(Nama="Budi", Umur=20, Prodi="TRK")
```

Praktikum 07 - Fungsi Decorator

```python
# kode untuk praktikum 07
```

```python
def decorator_sapa(fungsi):
  def wrapper():
    print("Sebelum fungsi dijalankan...")
    fungsi()
    print("Setelah fungsi dijalankan...")
  return wrapper

@decorator_sapa
def sapa():
  print("Halo, Mahasiswa Polines!")

sapa()
```

Tugas Fungsi Decorator

```python
# kode untuk tugas fungsi decorator
```

```python
def decorator_logger(fungsi):
  def wrapper():
    print("Memulai fungsi...")
    fungsi()
    print("Fungsi selesai dijalankan.")
  return wrapper

@decorator_logger
def halo():
  print("Halo, nama saya Vannisa")

halo()
```

Studi Kasus Nyata - Program Nilai

```python
# kode untuk studi kasus nyata - program nilai
```

```python
def input_nilai():
    nilai = []
    for i in range(1, 4):
        n = float(input(f"Masukkan nilai ke {i}: "))
        nilai.append(n)
    return nilai

def hitung_rata(nilai):
    return sum(nilai) / len(nilai)

def status_kelulusan(rata):
    if rata >= 60:
        return "Lulus"
    else:
        return "Tidak Lulus"

nilai = input_nilai()
rata = hitung_rata(nilai)
status = status_kelulusan(rata)

print(f"Rata-rata nilai: {rata:.2f}")
print(f"Status kelulusan: {status}")

nilai_akhir = lambda tugas, ujian: (0.4 * tugas) + (0.6 * ujian)
tugas = float(input("\nMasukkan nilai tugas: "))
ujian = float(input("Masukkan nilai ujian: "))
akhir = nilai_akhir(tugas, ujian)

print(f"Nilai akhir: {akhir:.2f}")
```
