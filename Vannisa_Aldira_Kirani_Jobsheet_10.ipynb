Praktikum 01 - Bubble Sort

```python
def bubble_sort(data):
  n = len(data)
  for i in range(n):
    for j in range(0, n - i - 1):
      if data[j] > data[j + 1]:
        data[j], data [j + 1] = data[j + 1], data[j]
  return data

angka = [64, 34, 25, 12, 22, 11, 90]
print("Data sebelum diurutkan:", angka)
print("Data setelah Bubble Sort:", bubble_sort(angka))
```

Praktikum 02 - Selection Sort

```python
def selection_sort(data):
  n = len(data)
  for i in range(n):
    min_index = i
    for j in range(i + 1, n):
      if data[j] < data[min_index]:
        min_index = j
    data[i], data[min_index] = data[min_index], data[i]
  return data

angka = [64, 25, 12, 22, 11]
print("Data sebelum diurutkan:", angka)
print("Data setelah Selection Sort:", selection_sort(angka))
```

Praktikum 03 - Insertion Sort

```python
def insertion_sort(data):
  n = len(data)
  for i in range(1, n):
    key = data[i]
    j = i - 1
    while j >= 0 and data[j] > key:
      data[j + 1] = data[j]
      j -= 1
    data[j + 1] = key
  return data

angka = [12, 11, 13, 5, 6]
print("Data sebelum diurutkan:", angka)
print("Data setelah Insertion Sort:", insertion_sort(angka))
```

Praktikum 04 - Studi Kasus: Mengurutkan Nilai Mahasiswa

```python
def bubble_sort_nilai(data):
  n = len(data)
  for i in range(n):
    for j in range(0, n - i - 1):
      if data[j][1] < data[j + 1][1]:
        data[j], data[j + 1] = data[j + 1], data[j]
  return data

mahasiswa = [("Andi", 80), ("Budi", 70), ("Cici", 90), ("Doni", 85)]
print("Data sebelum diurutkan:", mahasiswa)
print("Data setelah diurutkan (desc):", bubble_sort_nilai(mahasiswa))
```

Praktikum 06 - Sorting Data Barang (Nama & Harga)

```python
def bubble_sort_key(data, key_index=0, reverse=False):
  n = len(data)
  for i in range(n):
    for j in range(0, n - i - 1):
      if reverse:
        if data[j][key_index] < data[j + 1][key_index]:
          data[j], data[j + 1] = data[j + 1], data[j]
      else:
        if data[j][key_index] > data[j + 1][key_index]:
          data[j], data[j + 1] = data[j + 1], data[j]
  return data

barang = [("Buku", 15000), ("Pensil", 5000), ("Tas", 75000), ("Pulpen", 10000)]
print("Urut berdasarkan nama (asc):", bubble_sort_key(barang.copy(), 0))
print("Urut berdasarkan harga (desc):", bubble_sort_key(barang.copy(), 1, reverse=True))
```

Tugas Eksperimen Selection Sort

```python
def selection_sort_max(data):
  n = len(data)
  for i in range(n):
    max_index = 0
    for j in range(1, n - i):
      if data[j] > data[max_index]:
        max_index = j
    data[max_index], data[n - 1 - i] = data[n - 1 - i], data[max_index]
  return data

angka = [64, 25, 12, 22, 11]
print("Data sebelum diurutkan:", angka)
print("Data setelah Selection Sort:", selection_sort_max(angka))
```

Tugas Analisis Kompleksitas

```python
import random

def hitung_bubble(data):
  count = 0
  n = len(data)
  for i in range(n):
      for j in range(0, n - i - 1):
          count += 1
          if data[j] > data[j + 1]:
              data[j], data[j + 1] = data[j + 1], data[j]
  return count


def hitung_selection(data):
  count = 0
  n = len(data)
  for i in range(n):
      min_index = i
      for j in range(i + 1, n):
          count += 1
          if data[j] < data[min_index]:
              min_index = j
      data[i], data[min_index] = data[min_index], data[i]
  return count


def hitung_insertion(data):
  count = 0
  n = len(data)
  for i in range(1, n):
      key = data[i]
      j = i - 1
      while j >= 0 and data[j] > key:
          count += 1
          data[j + 1] = data[j]
          j -= 1
      count += 1
      data[j + 1] = key
  return count

ukuran = [10, 50, 100]

for n in ukuran:
  data = [random.randint(1, 1000) for _ in range(n)]

  print(f"\nJumlah Data: {n}")
  print("Bubble Sort   :", hitung_bubble(data.copy()))
  print("Selection Sort:", hitung_selection(data.copy()))
  print("Insertion Sort:", hitung_insertion(data.copy()))
```

```python
def bubble_sort_key(data, key_index=0, reverse=False):
  n = len(data)
  for i in range(n):
      for j in range(0, n - i - 1):
          if reverse:
              if data[j][key_index] < data[j + 1][key_index]:
                  data[j], data[j + 1] = data[j + 1], data[j]
          else:
              if data[j][key_index] > data[j + 1][key_index]:
                  data[j], data[j + 1] = data[j + 1], data[j]
  return data

barang = []
jumlah = int(input("Berapa banyak barang? "))

for i in range(jumlah):
  nama = input(f"Nama barang ke-{i+1}: ")
  harga = int(input("Harga: "))
  barang.append((nama, harga))

print("\nPilih cara pengurutan:")
print("1. Nama Ascending (A → Z)")
print("2. Nama Descending (Z → A)")
print("3. Harga Ascending (murah → mahal)")
print("4. Harga Descending (mahal → murah)")

pilih = int(input("Masukkan pilihan: "))

if pilih == 1:
    hasil = bubble_sort_key(barang.copy(), 0, reverse=False)
elif pilih == 2:
    hasil = bubble_sort_key(barang.copy(), 0, reverse=True)
elif pilih == 3:
    hasil = bubble_sort_key(barang.copy(), 1, reverse=False)
elif pilih == 4:
    hasil = bubble_sort_key(barang.copy(), 1, reverse=True)
else:
    hasil = "Pilihan tidak valid"

print("\nHasil pengurutan:")
print(hasil)
```
